## âš™ï¸ 5. Otomasyon Sistemi TanÄ±mla

10 hikayeyi sÄ±rayla iÅŸleyecek otomasyon sÄ±nÄ±fÄ±nÄ± tanÄ±mlÄ±yoruz.

````python
import sys
import json
import time
from datetime import datetime
from pathlib import Path
import traceback
import shutil

# ====================================================================
# EKSTRA KORUMA: pyttsx3 hatasÄ±nÄ± tamamen Ã¶nle
# ====================================================================
# Mock pyttsx3 modÃ¼lÃ¼ (eÄŸer hala import edilirse hata vermesin)
if 'pyttsx3' not in sys.modules:
    class MockPyttsx3:
        """Sahte pyttsx3 modÃ¼lÃ¼ - Colab iÃ§in"""
        @staticmethod
        def init():
            return None
    
    sys.modules['pyttsx3'] = MockPyttsx3()
    print("âœ… pyttsx3 mock modÃ¼lÃ¼ oluÅŸturuldu (gÃ¼venlik Ã¶nlemi)")

# Python path'e src klasÃ¶rÃ¼nÃ¼ ekle
sys.path.insert(0, '/content')

class VideoAutomation:
    """10 hikayeyi otomatik iÅŸleyen sistem - main.py'ye uygun"""
    
    def __init__(self):
        self.stories_dir = Path('/content/stories')
        self.output_dir = Path('/content/videos')
        self.drive_output = Path(OUTPUT_DIR)
        self.log_file = Path(f'{PROJECT_DIR}/automation_log.json')
        self.progress_file = Path(f'{PROJECT_DIR}/progress.json')
        
        # Ä°lerleme durumu
        self.progress = self.load_progress()
    
    def load_progress(self):
        """Ã–nceki ilerlemeyi yÃ¼kle (kesinti durumunda devam eder)"""
        if self.progress_file.exists():
            with open(self.progress_file, 'r', encoding='utf-8') as f:
                return json.load(f)
        return {
            'completed': [],
            'failed': [],
            'current': None,
            'total_processed': 0,
            'started_at': datetime.now().isoformat()
        }
    
    def save_progress(self):
        """Ä°lerlemeyi kaydet"""
        with open(self.progress_file, 'w', encoding='utf-8') as f:
            json.dump(self.progress, f, ensure_ascii=False, indent=2)
    
    def log_event(self, event_type, story_name, details):
        """OlaylarÄ± logla"""
        log_entry = {
            'timestamp': datetime.now().isoformat(),
            'type': event_type,
            'story': story_name,
            'details': details
        }
        
        logs = []
        if self.log_file.exists():
            with open(self.log_file, 'r', encoding='utf-8') as f:
                logs = json.load(f)
        
        logs.append(log_entry)
        
        with open(self.log_file, 'w', encoding='utf-8') as f:
            json.dump(logs, f, ensure_ascii=False, indent=2)
    
    def get_pending_stories(self):
        """Ä°ÅŸlenmemiÅŸ hikayeleri al"""
        all_stories = list(self.stories_dir.glob('*.txt'))
        completed = set(self.progress['completed'])
        pending = [s for s in all_stories if s.stem not in completed]
        return sorted(pending)
    
    def process_story(self, story_file):
        """Tek bir hikayeyi iÅŸle - TAMAMEN main.py'ye uygun"""
        story_name = story_file.stem
        
        print(f"\n{'='*70}")
        print(f"ðŸŽ¬ Ä°ÅžLENÄ°YOR: {story_name}")
        print(f"{'='*70}\n")
        
        self.progress['current'] = story_name
        self.save_progress()
        
        try:
            # ============================================================
            # 1. MODÃœL Ä°MPORT (main.py'deki gibi)
            # ============================================================
            from config.config import Config
            from src.story_processor import StoryProcessor
            from src.openai_tts_generator import OpenAITTSGenerator
            from src.multi_image_generator import MultiImageGenerator
            from src.character_manager import CharacterManager
            from src.video_creator import VideoCreator
            
            # ============================================================
            # 2. HÄ°KAYE Ä°ÅžLEME (main.py satÄ±r 95-103)
            # ============================================================
            print("ðŸ“– 1. Hikaye iÅŸleniyor...")
            
            story_processor = StoryProcessor(
                stories_dir=Config.STORIES_DIR,
                deepseek_api_key=Config.DEEPSEEK_API_KEY
            )
            
            story_text = story_processor.load_story(story_file.name)
            story_title = story_processor.get_story_title(story_text)
            scenes = story_processor.split_into_scenes(story_text)
            
            # AI baÅŸarÄ±sÄ±z olduysa None dÃ¶ner - iÅŸlemi durdur
            if scenes is None:
                raise Exception("DeepSeek AI analizi baÅŸarÄ±sÄ±z - lÃ¼tfen API ayarlarÄ±nÄ± kontrol edin")
            
            print(f"   âœ“ Hikaye: {story_title}")
            print(f"   âœ“ {len(scenes)} sahne oluÅŸturuldu")
            
            # ============================================================
            # 3. SES ÃœRETÄ°MÄ° - OpenAI TTS (main.py satÄ±r 107-140)
            # ============================================================
            print("\nðŸŽ¤ 2. Ses dosyalarÄ± oluÅŸturuluyor (OpenAI TTS-1 HD)...")
            
            # OpenAI TTS kullan
            if not os.environ.get('OPENAI_API_KEY'):
                raise Exception("OPENAI_API_KEY bulunamadÄ±! Colab Secrets'Ä± kontrol edin.")
            
            tts_generator = OpenAITTSGenerator(
                api_key=os.environ['OPENAI_API_KEY'],
                voice=Config.OPENAI_TTS_VOICE,
                language=Config.TTS_LANGUAGE,
                speed=Config.OPENAI_TTS_SPEED
            )
            
            audio_files = tts_generator.generate_story_audio(scenes, story_title)
            print(f"   âœ“ {len(audio_files)} ses dosyasÄ± oluÅŸturuldu")
            
            # ============================================================
            # 4. GÃ–RSEL ÃœRETÄ°MÄ° (main.py satÄ±r 145-170)
            # ============================================================
            print("\nðŸ–¼ï¸ 3. GÃ¶rseller oluÅŸturuluyor...")
            
            # Karakter yÃ¶neticisi
            char_manager = CharacterManager()
            
            # AI'dan karakterleri Ã§Ä±kar
            if hasattr(story_processor, 'ai_response') and story_processor.ai_response:
                characters = char_manager.extract_characters(story_processor.ai_response)
                if characters:
                    print(f"   â†’ Karakter tutarlÄ±lÄ±ÄŸÄ± aktif")
            
            # Multi-image generator (main.py'deki gibi)
            image_generator = MultiImageGenerator(
                hf_token=Config.HUGGINGFACE_API_KEY,
                replicate_token=Config.REPLICATE_API_KEY,
                use_free_alternative=Config.USE_FREE_IMAGES_ONLY
            )
            
            # Karakter yÃ¶neticisini baÄŸla
            image_generator.character_manager = char_manager
            
            # GÃ¶rselleri Ã¼ret
            image_files = image_generator.generate_story_images(scenes, story_title)
            print(f"   âœ“ {len(image_files)} gÃ¶rsel oluÅŸturuldu")
            
            # ============================================================
            # 5. VÄ°DEO OLUÅžTURMA (main.py satÄ±r 175-190)
            # ============================================================
            print("\nðŸŽ¥ 4. Video birleÅŸtiriliyor...")
            
            video_creator = VideoCreator(Config.VIDEOS_DIR)
            
            # ðŸŽ¯ VÄ°DEO HÄ°KAYE Ä°SMÄ°YLE KAYDEDÄ°LÄ°YOR!
            # VideoCreator artÄ±k story_title parametresini kullanarak
            # dosyayÄ± "Kibritci_Kiz.mp4" gibi anlamlÄ± isimle kaydediyor
            video_path = video_creator.create_story_video(
                scenes=scenes,
                image_files=image_files,
                audio_files=audio_files,
                story_title=story_title  # â† Bu sayede dosya adÄ± hikaye baÅŸlÄ±ÄŸÄ± olacak
            )
            
            # ============================================================
            # 6. GOOGLE DRIVE'A KOPYALA
            # ============================================================
            if video_path and Path(video_path).exists():
                drive_path = self.drive_output / Path(video_path).name
                shutil.copy2(video_path, drive_path)
                
                file_size = Path(video_path).stat().st_size / (1024*1024)
                
                # Video bilgileri
                video_info = video_creator.get_video_info(video_path)
                
                self.progress['completed'].append(story_name)
                self.log_event('SUCCESS', story_name, {
                    'video_path': str(drive_path),
                    'size_mb': round(file_size, 2),
                    'scenes': len(scenes),
                    'duration': video_info.get('duration', 0)
                })
                
                print(f"\nâœ… BAÅžARILI: {story_name}")
                print(f"ðŸ“ Video dosyasÄ±: {Path(video_path).name}")  # Hikaye ismiyle kaydedilmiÅŸ dosya
                print(f"ðŸ“‚ Drive yolu: {drive_path}")
                print(f"ðŸ“Š Boyut: {file_size:.1f} MB")
                print(f"â±ï¸  SÃ¼re: {video_info.get('duration', 0):.1f} saniye")
                
                # Temizlik
                try:
                    video_creator.cleanup_temp_files()
                except:
                    pass
                
                return True
            else:
                raise Exception("Video dosyasÄ± oluÅŸturulamadÄ±")
                
        except Exception as e:
            error_msg = str(e)
            error_trace = traceback.format_exc()
            
            self.progress['failed'].append({
                'story': story_name,
                'error': error_msg,
                'timestamp': datetime.now().isoformat()
            })
            self.log_event('ERROR', story_name, {
                'error': error_msg,
                'traceback': error_trace
            })
            
            print(f"\nâŒ HATA: {story_name}")
            print(f"Hata mesajÄ±: {error_msg}")
            print(f"\nDetaylÄ± hata:")
            print(error_trace)
            return False
        
        finally:
            self.progress['current'] = None
            self.progress['total_processed'] += 1
            self.save_progress()
    
    def run(self):
        """TÃ¼m hikayeleri iÅŸle"""
        start_time = time.time()
        pending = self.get_pending_stories()
        
        if not pending:
            print("\nâš ï¸ Ä°ÅŸlenecek hikaye bulunamadÄ±!")
            print("TÃ¼m hikayeler zaten iÅŸlenmiÅŸ olabilir.")
            print(f"\nTamamlanan: {self.progress['completed']}")
            return self.progress
        
        print(f"\n{'='*70}")
        print(f"ðŸš€ OTOMATÄ°K Ä°ÅžLEME BAÅžLIYOR")
        print(f"{'='*70}")
        print(f"ðŸ“Š Toplam hikaye: {len(pending)}")
        print(f"â±ï¸  BaÅŸlangÄ±Ã§: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
        
        if self.progress['completed']:
            print(f"âœ… Daha Ã¶nce tamamlanan: {len(self.progress['completed'])}")
        
        print(f"{'='*70}\n")
        
        success_count = 0
        fail_count = 0
        
        for i, story_file in enumerate(pending, 1):
            print(f"\nâ”Œ{'â”€'*68}â”")
            print(f"â”‚ Ä°LERLEME: [{i}/{len(pending)}] {story_file.stem:^50} â”‚")
            print(f"â””{'â”€'*68}â”˜")
            
            result = self.process_story(story_file)
            if result:
                success_count += 1
            else:
                fail_count += 1
            
            # Her hikaye sonrasÄ± kÄ±sa ara (API rate limit iÃ§in)
            if i < len(pending):
                print("\nâ³ 30 saniye bekleniyor (API korumasÄ±)...")
                time.sleep(30)
        
        # Ã–zet rapor
        elapsed = time.time() - start_time
        hours = int(elapsed // 3600)
        minutes = int((elapsed % 3600) // 60)
        
        print(f"\n\n{'='*70}")
        print(f"ðŸŽ‰ TÃœM Ä°ÅžLEMLER TAMAMLANDI!")
        print(f"{'='*70}")
        print(f"âœ… BaÅŸarÄ±lÄ±: {success_count}")
        print(f"âŒ HatalÄ±: {fail_count}")
        print(f"ðŸ“Š Toplam tamamlanan: {len(self.progress['completed'])}")
        print(f"â±ï¸  SÃ¼re: {hours} saat {minutes} dakika")
        print(f"ðŸ“ Ã‡Ä±ktÄ±: {self.drive_output}")
        print(f"ðŸŽ¯ Video dosyalarÄ± hikaye isimleriyle kaydedildi!")
        print(f"{'='*70}\n")
        
        if self.progress['failed']:
            print("\nâš ï¸ HATALI HÄ°KAYELER:")
            for fail in self.progress['failed']:
                print(f"  - {fail['story']}: {fail['error']}")
        
        return self.progress

print("âœ… VideoAutomation sÄ±nÄ±fÄ± tanÄ±mlandÄ±! (main.py'ye %100 uyumlu)")
print("ðŸ›¡ï¸  pyttsx3 korumasÄ± aktif!")
print("ðŸŽ¯ Video dosyalarÄ± artÄ±k hikaye isimleriyle kaydediliyor!")
````
